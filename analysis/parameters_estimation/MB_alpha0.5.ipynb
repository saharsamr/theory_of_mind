```python
from analysis.correctness_checker.behavior_analysis.MB_agent.plot import plot_coeffs
import matplotlib.pyplot as plt
```

```python
def plots(alpha, beta):
    
    model, result = plot_coeffs('analysis/correctness_checker/behavior_analysis/data/MB/', alpha, beta)
    
    common_reward_mean = result.fe_mean[1]
    reward_prob_mean = result.fe_mean[2]
    interaction_mean = result.fe_mean[3]

    common_reward_sd = result.fe_sd[1]
    reward_prob_sd = result.fe_sd[2]
    interaction_sd = result.fe_sd[3]
    
    xs = [1, 2, 3]
    plt.errorbar(xs, [common_reward_mean, reward_prob_mean, interaction_mean], [common_reward_sd, reward_prob_sd, interaction_sd], fmt='o')
    plt.xticks(xs, ['common_reward', 'reward_prob', 'interaction'])
    plt.show()  
```

```python
alpha = 0.5
beta = 1.0
plots(alpha, beta)
```
